## 2. feature 추가하기

```python
# Dst Port가 3128인지 여부를 bool 값으로 추가
train_df['Is Port 3128'] = train_df['Dst Port'] == 3128
valid_df['Is Port 3128'] = valid_df['Dst Port'] == 3128
```

```python
train_df.columns
```

```python
col = ['Src Port', 'Dst Port','Protocol', 'Flow Duration'
       'Total Length of Fwd Packet', 'Total Length of Bwd Packet',
       'Fwd Header Length', 'Bwd Header Length',
       'Total Fwd Packet', 'Total Bwd packets', 
       'Average Packet Size']
```

도메인 지식을 활용하여 feature를 자유롭게 추가하시면 됩니다.

```python
train_expt = train_df[col]
#train_expt = pd.get_dummies(train_expt, columns=['Protocol'], prefix='Protocol')
train_list = train_expt.values.tolist()

valid_expt = valid_df[col]
#valid_expt = pd.get_dummies(valid_expt, columns=['Protocol'], prefix='Protocol')
valid_list = valid_expt.values.tolist()
```

```python
train_y = [0 if i == "Benign_IP" else 1 for i in train_df['Label']]
valid_y = [0 if i == "Benign_IP" else 1 for i in valid_df['Label']]
```

### 스케일러 적용

```python
# standardScaler 
from sklearn.preprocessing import StandardScaler

normalizer = StandardScaler()
train_X = []
valid_X = []

normalizer.fit(train_list)

train_X = normalizer.transform(train_list)
valid_X = normalizer.transform(valid_list)
```

```python
#학습을 위한 데이터 타입 변환

train_X = train_X.astype(np.float32)
valid_X = valid_X.astype(np.float32)
train_y = np.array(train_y, dtype=np.float32)
valid_y = np.array(valid_y, dtype=np.float32)
```

# 3. 모델 학습 

```python
from xgboost import XGBClassifier
xg = XGBClassifier(n_estimators=100, learning_rate=0.1, objective='binary:logistic')
xg.fit(train_X, train_y)
xg_pred = xg.predict(valid_X)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# 성능 평가를 위해 예측 값을 반올림
xg_round_valid_y = valid_y.astype("int")
xg_round_y_pred = np.round(xg_pred)
print("confusion_matrix\n", confusion_matrix(xg_round_valid_y, xg_round_y_pred))
print(f"accuracy\t: {accuracy_score(xg_round_valid_y, xg_round_y_pred): .5f}")
print(f"precision\t: {precision_score(xg_round_valid_y, xg_round_y_pred): .5f}")
print(f"recall\t\t: {recall_score(xg_round_valid_y, xg_round_y_pred): .5f}")
print(f"F-1\t\t: {f1_score(xg_round_valid_y, xg_round_y_pred): .5f}")
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(n_estimators=100, random_state=108) # randomstate는 108로 고정 
rfc.fit(train_X, train_y)
y_pred = rfc.predict(valid_X)
```

플로우 기반 이상탐지 성능평가

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# 성능 평가를 위해 예측 값을 반올림
round_valid_y = valid_y.astype("int")
round_y_pred = np.round(y_pred)
print("confusion_matrix\n", confusion_matrix(round_valid_y, round_y_pred))
print(f"accuracy\t: {accuracy_score(round_valid_y, round_y_pred): .5f}")
print(f"precision\t: {precision_score(round_valid_y, round_y_pred): .5f}")
print(f"recall\t\t: {recall_score(round_valid_y, round_y_pred): .5f}")
print(f"F-1\t\t: {f1_score(round_valid_y, round_y_pred): .5f}")
```

# 4. IP voting

실제 성능평가는 IP기준으로 해야하기 때문에 flow 기반 예측결과를 이용하여 IP votiong을 진행합니다.

```python
# 정상 외부 IP set
outer_benignIP_set = open_pickle("./outer_benign_IP_train.pkl")
# 악성 외부 IP set
outer_malIP_set = open_pickle("./outer_mal_IP_train.pkl")
```

```python
#외부 IP의 악성, 정상 개수 확인
len(outer_benignIP_set), len(outer_malIP_set)
```

```python
# Problem #3을 위해 dictionary형태로 변경
labelIP_dict = dict()

for ip in outer_benignIP_set:
    labelIP_dict[ip] = 0
for ip in outer_malIP_set:
    labelIP_dict[ip] = 1
```

```python
xg_predictIP_dict = {}

for idx, data in enumerate(valid_df.values):
    if data[0] in labelIP_dict:
        if data[0] not in xg_predictIP_dict:
            xg_predictIP_dict[data[0]] = 0

        if xg_pred[idx] == 1:
            xg_predictIP_dict[data[0]] = 1

    else:
        if data[2] not in xg_predictIP_dict:
            xg_predictIP_dict[data[2]] = 0

        if xg_pred[idx] == 1:
            xg_predictIP_dict[data[2]] = 1
```

```python
xg_pred_IP = []
xg_label_IP = []
xg_TP_IP = []
xg_TN_IP = []
xg_FP_IP = []
xg_FN_IP = []


for k in xg_predictIP_dict.keys():
    pred_IP = xg_predictIP_dict[k]
    label_IP = labelIP_dict[k]
    if pred_IP == 1 and label_IP == 1:
        xg_TP_IP.append(k)
    elif pred_IP == 0 and label_IP == 0:
        xg_TN_IP.append(k)
    elif pred_IP == 1 and label_IP == 0:
        xg_FP_IP.append(k)
    elif pred_IP == 0 and label_IP == 1:
        xg_FN_IP.append(k)
    
    xg_pred_IP.append(pred_IP)
    xg_label_IP.append(label_IP)

print("confusion_matrix\n", confusion_matrix(xg_label_IP, xg_pred_IP))
print(f"accuracy\t: {accuracy_score(xg_label_IP, xg_pred_IP): .5f}")
print(f"precision\t: {precision_score(xg_label_IP, xg_pred_IP): .5f}")
print(f"recall\t\t: {recall_score(xg_label_IP, xg_pred_IP): .5f}")
print(f"F-1\t\t: {f1_score(xg_label_IP, xg_pred_IP): .5f}")
```

```python
group_valid_df = valid_df.groupby('Src IP')
```

```python
tp_data = valid_df[valid_df['Src IP'].isin(xg_TP_IP)]
fp_data = valid_df[valid_df['Src IP'].isin(xg_FP_IP)]
tn_data = valid_df[valid_df['Src IP'].isin(xg_TN_IP)]
fn_data = valid_df[valid_df['Src IP'].isin(xg_FN_IP)]

comparison = 'Average Packet Size'
tp_data[comparison].mean(), fp_data[comparison].mean(), tn_data[comparison].mean(), fn_data[comparison].mean()
```

```python
predictIP_dict = {}

for idx, data in enumerate(valid_df.values):
    if data[0] in labelIP_dict:
        if data[0] not in predictIP_dict:
            predictIP_dict[data[0]] = 0

        if y_pred[idx] == 1:
            predictIP_dict[data[0]] = 1

    else:
        if data[2] not in predictIP_dict:
            predictIP_dict[data[2]] = 0

        if y_pred[idx] == 1:
            predictIP_dict[data[2]] = 1
```

```python
pred_IP = []
label_IP = []
TP_IP = []
TN_IP = []
FP_IP = []
FN_IP = []

for k in predictIP_dict.keys():
    pred = 1 if predictIP_dict[k] == 1 else 0
    label = labelIP_dict[k]
    if pred == 1 and label == 1:
        TP_IP.append(k)
    elif pred == 0 and label == 0:
        TN_IP.append(k)
    elif pred == 1 and label == 0:
        FP_IP.append(k)
    elif pred == 0 and label == 1:
        FN_IP.append(k)
    pred_IP.append(1 if predictIP_dict[k] == 1 else 0)
    label_IP.append(labelIP_dict[k])

print("confusion_matrix\n", confusion_matrix(label_IP, pred_IP))
print(f"accuracy\t: {accuracy_score(label_IP, pred_IP): .5f}")
print(f"precision\t: {precision_score(label_IP, pred_IP): .5f}")
print(f"recall\t\t: {recall_score(label_IP, pred_IP): .5f}")
print(f"F-1\t\t: {f1_score(label_IP, pred_IP): .5f}")
```
